```python
max_par = (0, 1)
max_cov = cov_matrix.iloc[0, 1]
for i in range(d):
    for j in range(d):
        if i < j and cov_matrix.iloc[i, j] > max_cov:
            max_par = (i, j)
            max_cov = cov_matrix.iloc[i, j]

print(max_par)
print(max_cov)            
```
