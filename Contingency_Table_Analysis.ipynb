```python
def contingency_table_analysis(cat1, cat2):
    if not isinstance(df[cat1].dtype, pd.CategoricalDtype):
        df[cat1] = df[cat1].astype('category')
    if not isinstance(df[cat2].dtype, pd.CategoricalDtype):
        df[cat2] = df[cat2].astype('category')

    contingency_table = df.pivot_table(
        values='age',
        columns=cat2,
        index=cat1,
        aggfunc='count',
        margins=True,
        margins_name='Total'
    )

    m1, m2 = contingency_table.shape
    m1 -= 1
    m2 -= 1
    q = (m1-1)*(m2-1)
    
    expected_frequency_table = contingency_table.apply(
        lambda x : x[-1] * contingency_table.iloc[:, -1] / contingency_table.iloc[-1, -1]
    )

    chi2_value = ((expected_frequency_table - contingency_table)**2 / expected_frequency_table).sum().sum()

    chi2_stats = stats.chi2(q)

    p_value = 1 - chi2_stats.cdf(chi2_value)

    return chi2_value, p_value
```
