```python
import networkx as nx
```

```python
G = nx.Graph()

H = nx.DiGraph()
```

```python
G.add_node('a')

nodes_to_add = ['b', 'c', 'd']
G.add_nodes_from(nodes_to_add)


nx.draw(G, with_labels=True)
```

```python
# add edge from 'a', 'b'
G.add_edge('a', 'b')

nx.draw(G, with_labels=True)
```

add eges from a sequence

```python
edges_to_add = [('a', 'c'), ('b', 'c'), ('c', 'd')]
G.add_edges_from(edges_to_add)


nx.draw(G, with_labels=True)
```

```python
nx.draw(G, with_labels=True,
        node_color='r',
        node_size=1600,
        font_color='white',
        font_size=16,
        )
```

Node degree

```python
G.degree('c')
```

```python
def get_leaves(G):
    lst = []
    for x in G.nodes:
        if G.degree(x) == 1:
            lst.append(x)

    return lst        
```

```python
get_leaves(G)
```
